```python
cat_col = ['job','marital','education','default','balance','housing','loan','contact','day','month','duration','campaign','pdays','previous','poutcome','deposit']
fig = plt.figure(figsize=(20,20))
for i,col in enumerate(cat_col):
    ax = fig.add_subplot(4 , 3 , i+1)
    sns.countplot(data=data , x=col )
```

```python
for i in cat_col[:-1]:
    sns.countplot(data=data , x=col , hue='deposit')
    plt.xticks(rotation = 45)
    plt.title(f"{i} vs Deposit")
```
